```python
from tkinter import*
from tkinter.ttk import Radiobutton
qstns={'questions':['1.Who is the prime minister of india?','2.Total number of districts in Kerala?','3.Which is the smallest state in india?','4.Total number of states in india?','5.Which is the largest state in india?'],'options':[['H.D.Gawda','Chandra Shekar','Rajiv Gandhi','Narendra Modi'],[14,13,12,11],['Kerala','Goa','Tamil Nadu','Karnataka'],[27,28,29,30],['MP','Maharashtra','AP','Rajasthan']]}


window=Tk()
window.title('QUIZ BOARD')
window.geometry('300x900')

entry=Entry(window,width=30)
entry.place(x=50,y=560)

lbl1=Label(window,text=qstns['questions'][0])
lbl1.place(x=40,y=0)
lbl2=Label(window,text=qstns['questions'][1])
lbl2.place(x=40,y=100)
lbl3=Label(window,text=qstns['questions'][2])
lbl3.place(x=40,y=200)
lbl4=Label(window,text=qstns['questions'][3])
lbl4.place(x=40,y=300)
lbl5=Label(window,text=qstns['questions'][4])
lbl5.place(x=40,y=400)

selected1=StringVar()
rad1=Radiobutton(window,text=qstns['options'][0][0],value='HD.Gawda',variable=selected1)
rad1.place(x=40,y=20)
rad1=Radiobutton(window,text=qstns['options'][0][1],value='Chandra Shekar',variable=selected1)
rad1.place(x=40,y=40)
rad1=Radiobutton(window,text=qstns['options'][0][2],value='Rajiv Gandhi',variable=selected1)
rad1.place(x=40,y=60)
rad1=Radiobutton(window,text=qstns['options'][0][3],value='Narendra Modi',variable=selected1)
rad1.place(x=40,y=80)

selected2=IntVar()
rad1=Radiobutton(window,text=qstns['options'][1][0],value=14,variable=selected2)
rad1.place(x=40,y=120)
rad1=Radiobutton(window,text=qstns['options'][1][1],value=13,variable=selected2)
rad1.place(x=40,y=140)
rad1=Radiobutton(window,text=qstns['options'][1][2],value=12,variable=selected2)
rad1.place(x=40,y=160)
rad1=Radiobutton(window,text=qstns['options'][1][3],value=11,variable=selected2)
rad1.place(x=40,y=180)

selected3=StringVar()
rad1=Radiobutton(window,text=qstns['options'][2][0],value='Kerala',variable=selected3)
rad1.place(x=40,y=220)
rad1=Radiobutton(window,text=qstns['options'][2][1],value='Goa',variable=selected3)
rad1.place(x=40,y=240)
rad1=Radiobutton(window,text=qstns['options'][2][2],value='Tamil Nadu',variable=selected3)
rad1.place(x=40,y=260)
rad1=Radiobutton(window,text=qstns['options'][2][3],value='Karnataka',variable=selected3)
rad1.place(x=40,y=280)

selected4=IntVar()
rad1=Radiobutton(window,text=qstns['options'][3][0],value=27,variable=selected4)
rad1.place(x=40,y=320)
rad1=Radiobutton(window,text=qstns['options'][3][1],value=28,variable=selected4)
rad1.place(x=40,y=340)
rad1=Radiobutton(window,text=qstns['options'][3][2],value=29,variable=selected4)
rad1.place(x=40,y=360)
rad1=Radiobutton(window,text=qstns['options'][3][3],value=30,variable=selected4)
rad1.place(x=40,y=380)

selected5=StringVar()
rad1=Radiobutton(window,text=qstns['options'][4][0],value='MP',variable=selected5)
rad1.place(x=40,y=420)
rad1=Radiobutton(window,text=qstns['options'][4][1],value='Maharashtra',variable=selected5)
rad1.place(x=40,y=440)
rad1=Radiobutton(window,text=qstns['options'][4][2],value='AP',variable=selected5)
rad1.place(x=40,y=460)
rad1=Radiobutton(window,text=qstns['options'][4][3],value='Rajasthan',variable=selected5)
rad1.place(x=40,y=480)


bt1=Button(window,text='RESULT',bg='blue',fg='white',command=result)
bt1.place(x=100,y=520)

a=['Narendra Modi',14,'Goa',29,'Rajasthan']

def result():
    try:
        score=0
        if a[0]==selected1.get():
            score=score+1
        if a[1]==selected2.get():
            score=score+1
        if a[2]==selected3.get():
            score=score+1
        if a[3]==selected4.get():
            score=score+1
        if a[4]==selected5.get():
            score=score+1
        entry.insert(0,score)
    finally:
        entry.delete(0,END)
        entry.insert(0,score)

    
    
window.mainloop()
```
